### Analysis

- The study covered 249 mice being treated by 10 different drug regimens over a 45 day period. There was a 51/49 Male/Female ratio of mice participating in the study.

- Capomulin and Ramicane had the highest number of data points across mice and timepoints. These two drug regimes were also the most successful over the 45 days in reducing tumor volumes.

- Upon reviewing the Capomulin treated mice data, there is a strong correlation (0.84) between the average tumor volume and the weight.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mousestudy_combined = pd.merge(study_results,mouse_metadata,how="outer",on="Mouse ID")

# Display the data table for preview
mousestudy_combined.head()
```

```python
# Checking the number of mice.
mousestudy_combined["Mouse ID"].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_duplicate = mousestudy_combined.loc[mousestudy_combined[["Mouse ID","Timepoint"]].duplicated(),["Mouse ID"]]
mouse_duplicate["Mouse ID"].info
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouse_duplicates = mousestudy_combined[mousestudy_combined[["Mouse ID","Timepoint"]].duplicated()]
mouse_duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mousestudy_combined.drop_duplicates(subset=["Mouse ID","Timepoint"])
```

```python
# Checking the number of mice in the clean DataFrame.
mousestudy_combined["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
mousestudy_group = mousestudy_combined.groupby("Drug Regimen")

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mousestudy_statistics=pd.DataFrame()
#Mean Tumor Volume
Mean_Tumor_Volume = mousestudy_group["Tumor Volume (mm3)"].mean()
mousestudy_statistics["Mean Tumor Volume"]=Mean_Tumor_Volume.map("{:,.2f}".format)
#Median Tumor Volume
Median_Tumor_Volume = mousestudy_group["Tumor Volume (mm3)"].median()
mousestudy_statistics["Median Tumor Volume"]=Median_Tumor_Volume.map("{:,.2f}".format)
# Tumor Volume Variance
Tumor_Volume_Variance = mousestudy_group ["Tumor Volume (mm3)"].var()
mousestudy_statistics["Tumor Volume Variance "]= Tumor_Volume_Variance.map("{:,.2f}".format)
# Tumor Volume Std. Dev.
Tumor_Volume_Std_Dev= mousestudy_group ["Tumor Volume (mm3)"].std()
mousestudy_statistics["Tumor Volume Std. Dev."]= Tumor_Volume_Std_Dev.map("{:,.2f}".format)
# Tumor Volume Std. Err.
Tumor_Volume_Std_Err= mousestudy_group ["Tumor Volume (mm3)"].sem()
mousestudy_statistics["Tumor Volume Std. Err."] = Tumor_Volume_Std_Err.map("{:,.2f}".format)

# Assemble the resulting series into a single summary DataFrame.
mousestudy_statistics
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
mousestudy_group_agg = mousestudy_group.agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
mousestudy_group_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
treatments_list = ("Capomulin", "Ramicane", "Infubinol", "Ceftamin")
# Start by getting the last (greatest) timepoint for each mouse
LastTP = mousestudy_combined.groupby('Mouse ID').max()['Timepoint']
LastTP_vol = pd.DataFrame(LastTP)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
LastTP_merge = pd.merge(LastTP_vol, mousestudy_combined, on=("Timepoint","Mouse ID"),how="inner")
LastTP_treatments = LastTP_merge[LastTP_merge["Drug Regimen"].isin(treatments_list)]
LastTP_treatments
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments_list #see above   

# Create empty list to fill with tumor vol data (for plotting)
Tumor_Vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for treatment in treatments_list:    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    Tumorvolumes = LastTP_treatments.loc[LastTP_treatments["Drug Regimen"]==treatment,"Tumor Volume (mm3)"]
    
    
    # add subset 
    Tumor_Vol.append(Tumorvolumes)
    
    # Determine outliers using upper and lower bounds
    quartiles = Tumorvolumes.quantile([0.25,0.5,0.75]).round(2)
    lowerq = quartiles[0.25].round(2)
    upperq = quartiles[0.75].round(2)
    iqr = round(upperq-lowerq,2)
    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)
    
    print(f"{treatment}'s potential outliers: Values below {lower_bound} and above {upper_bound}.")
    print(f"------------------------------------------------------------")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
    
fig, ax = plt.subplots()

ax.set_xlabel("Treatment Regimen")
ax.set_ylabel("Tumor Volume")
ax.set_title("Final Tumor Volume per Drug Regimen")
ax.boxplot(Tumor_Vol,0,"rd")
ax.set_xticklabels(treatments_list)
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
Capomulin_df = mousestudy_combined.loc[(mousestudy_combined["Drug Regimen"]=="Capomulin")]
Capomulin_AverageV= Capomulin_df.groupby(["Mouse ID"]).mean()
Capomulin_AverageV.plot.scatter(x="Weight (g)",y="Tumor Volume (mm3)",ylabel = "Average Tumor Volume (mm3)",title = 'Comparison of Average Tumor Volume (mm3) to Weight (g) of Capomulin treatmeant')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
corr_coeff = round(st.pearsonr(Capomulin_AverageV["Weight (g)"],Capomulin_AverageV["Tumor Volume (mm3)"])[0],2)
print(f"The correlation between mouse weight and the average tumor volume is {corr_coeff}")
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(Capomulin_AverageV["Weight (g)"], Capomulin_AverageV["Tumor Volume (mm3)"])
regress_values = Capomulin_AverageV["Weight (g)"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))

# Create Plot
plt.scatter(Capomulin_AverageV["Weight (g)"],Capomulin_AverageV["Tumor Volume (mm3)"])
plt.plot(Capomulin_AverageV["Weight (g)"],regress_values,"r-")

# Label plot and annotate the line equation
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

# Show plot
plt.show()
```
